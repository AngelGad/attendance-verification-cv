```python
import re
import pandas as pd
import csv
import cv2
import os
from os import path
import pickle
import face_recognition
from imutils import paths
import numpy as np
import datetime
from datetime import datetime,date
import tkinter as tk
from tkinter import *
import tkinter.ttk as ttk
import csv
from tkinter import messagebox
from tkinter import messagebox
import warnings
global pathm
warnings.filterwarnings('ignore')


# designing the welcome interface 
ws = tk.Tk()
ws.title('Registration') # the name of the page
ws.geometry('400x500')  # the size of page   
style = ttk.Style() 
style.theme_use('classic') # using classic styling
dd = pd.Timestamp(datetime.today()) # getting today's date 
dd = dd.day_name() # today's name
today = date.today() 
main_day = today.strftime("%m/%d/%y") # format to how the date should display 
student_image_path = []

          
def register():
    #setting up the  first page of the GUI which is to create student profile
    profile = Toplevel(ws) # for the button to be inside the main welcome page
    profile.title('Create a Profile') # the name of the page to create 
    profile.geometry('500x500') # the size of the interface
    menubar = Menu(ws) # for other menu options on the page 
    profile.config(menu=menubar) # to configure the menu bar inside the registration page  
    homemenu= Menu(menubar,tearoff=0)
    homemenu.add_command(command=profile.destroy,label='Exit') # just one menubar which is to exit the page
    menubar.add_cascade(label="File",menu=homemenu) # the name of the menu option is file 
   
    dd = pd.Timestamp(datetime.today()) # to give us today's date
    dd = dd.day_name() # giving us the today's name
    today = date.today() 
    main_day = today.strftime("%m/%d/%y") # the format of how the date should be arranged 
    
    # setting up the entry boxes for users to input their details 
    firstname = Label(profile, text ='First Name:', font = ('Times', 15),  fg = 'black')   # Label is to write text for description of the entry box 
    firstname.grid(row=1, column = 0) # the position the lebel 'first name should take'
    firstnamey = Entry(profile, width = 20) # enty box where user can input the first name 
    firstnamey.grid(row = 1, column = 1) # the position is should take 
    middlename = Label(profile, text ='Middle Name:', font = ('Times', 15),  fg = 'black')
    middlename.grid(row=2, column = 0)
    middlenamey = Entry(profile, width = 20)      
    middlenamey.grid(row = 2, column = 1)
    lastname = Label(profile, text ='Last Name:', font = ('Times', 15),  fg = 'black')
    lastname.grid(row=3, column = 0)
    lastnamey = Entry(profile, width = 20,)
    lastnamey.grid(row = 3, column = 1)
    gender= Label(profile,text= 'Gender:', font= ('Times',15),  fg = 'black')
    gender.grid(row=4,column=0)
    Var= StringVar()
    Radiobutton0= ttk.Radiobutton(profile,text='Male',variable=Var,value='Male').grid(row=4,column=1)
    Radiobutton= ttk.Radiobutton(profile,text='Female', variable=Var,value='Female').grid(row=4,column=2)
    email = Label(profile,text = 'Email:', font = ('Times',15),  fg = 'black')
    email.grid (row=5,column=0)
    emaily = Entry(profile,width = 30)
    emaily.grid(row =5,column=1)
    tp = Label(profile, text = 'Telephone Number:', font = ('Times',15),  fg = 'black')
    tp.grid(row=7,column=0)
    tpy=Entry(profile,width=30)
    tpy.grid(row=7,column=1)
    add = Label(profile, text = 'Address:', font = ('Times',15),  fg = 'black')
    add.grid(row=8,column=0)
    add1=Entry(profile,width=30)
    add1.grid(row=8,column=1)
    course = Label(profile, text = 'Course:', font = ('Times',15),  fg = 'black')
    course.grid(row=9,column=0)
    course1=Entry(profile,width=30)
    course1.grid(row=9,column=1)   
    dept = Label(profile, text = 'Department:', font = ('Times',15),  fg = 'black')
    dept.grid(row=10,column=0)
    dept1=Entry(profile,width=30)
    dept1.grid(row=10,column=1)    

    # capture the user face
    def camera():
        if firstnamey.get()=='' or lastnamey.get()=='': # if first name or last name is empty i.e if any of the two is empty, warning box would be shown to the user to input it 
            messagebox.showerror(profile, "Field can't be empty ",icon='warning')
            return
        name = (firstnamey.get().upper()) # getting the first name only to create folder for the user in other to save his/her picture, because each user will have their folder
        path1 = f"C:/Users/ANGEL GAD/Desktop/Azeez Shittu Pro/Student Images/{name}" #generating new folder name fold the user with his or her name 
        try:
            os.mkdir(path1) # create the folder in the local directory
        except OSError:
            messagebox.showinfo(profile,"Creation of the directory %s failed" % path1,icon='info') # if the folder is not able to create, there show be a notification informing the user to satrt again
        else:
            messagebox.showinfo(profile,"Successfully created the directory %s " % path1,icon='info')  # the folder is created successfully 
            messagebox.showinfo(profile, 'Press s to stop once box has been detected in your Face', icon='info') # notification for the command button to use while taking the picture
            cascade = cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml') # loading the haarcascade pre-train model
            video = cv2.VideoCapture(0) # on the camera automatically 
            while 1: # to be in the loop till the operation is done.
                # cont = 1
                ret,frame = video.read() # the vidoe return two values which are 1. the condition of the video. i.e if the video has been intialized (this return true or false) and 2. the frame of the image getting (it returns array datatype) 
                gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY) # for faster recognition while marking the attendance we turn it to gray because the colour alway have effect on the image while recogizing it  
                faces = cascade.detectMultiScale(gray,scaleFactor= 1.2,minNeighbors=7) # scale factor: this is to specify the size of the image and the minneigbour is the number of neigbors the rectangle should have
                for x,y,w,h in faces: # the face return the y and x axis, height and width  
                    frame = cv2.rectangle(gray,(x,y),(x+w,y+h),(128,128,128),3)  # creating a tringle box from the face according to the size we want 
                    pathm = path.join(path1, f'{name}.jpg') # creating the path for the image gotten from the rectangle
                    cv2.imwrite(pathm,frame) # downloading the image
                    student_image_path.append(pathm) # generating the image path as to save it also tho the student profile database
                cv2.imshow('Studet Registration portal',frame) # Frame name
                key = cv2.waitKey(15) 
                if key == ord('s'): # command key
                    break
            video.release()
            cv2.destroyAllWindows()
            # Image enrolment 
        for image_path in list(paths.list_images('Student Images')): #checking the number of images that is in the folder
            known_names = [] # empty list to append the name of the image   
            known_encodings =[] # empty list to append the code that will be generated for the image 
            name = image_path.split('\\')[1] # only one \ is needed from the image path directory  
            img = cv2.imread(image_path) # reading the image 
            rgb_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) # converting the image from BGR2RGB due to how the fact that open cv use BGR image format 
            faces = face_recognition.face_locations(rgb_img,model='hog') #using hog model which means Histogram of Oriented Gradients 
            encodings = face_recognition.face_encodings(rgb_img,faces) # generating the coded value
            for encoding in encodings:
                known_names.append(name)
                known_encodings.append(encoding)
            data = {'Names':known_names,'Encodings':known_encodings} 
            with open('Student_encoder_new','wb') as f: 
                    f.write(pickle.dumps(data)) # saving as pickle

    submit1 = Button(profile, text = 'Take Photo', command = camera,font=('Times', 12), bg= 'black', fg= 'blue') # the button for taking the picture
    submit1.grid(row =11, column=1)
        
    def submit():
        if student_image_path:
            # getting all the information
            if str(firstnamey.get())==''  or str(lastnamey.get())=='' or str(emaily.get())=='' or str(tpy.get())=='' or str(add1)==''  or str(Var.get())=='' or str(dept1.get())=='':
                messagebox.showerror(profile, "Field can't be empty",icon='warning')
                return
            # check if there is @ and . in the email
            if re.search ('[@.]',str(emaily.get()))!=None:
                print ('email ok')
            else:
                messagebox.showerror(profile, 'invalid Email',icon='warning')
                return
            # check if there input is in digit
            if re.search('[0-9]',str(tpy.get()))!=None:
                print('Telephone ok')
            else:
                messagebox.showerror(profile,'Invalid Telephone Number',icon='warning')
                return
            name = (firstnamey.get().upper())
            file_name = 'Student Profile.csv' # the name of the database
            path1 =f"C:/Users/ANGEL GAD/Desktop/Azeez Shittu Pro/Student Images/{name}"
            # appending everythig to the csv file
            if os.path.exists(file_name):
                df = pd.read_csv(file_name)
                data = {
                    'First Name': firstnamey.get().upper(),
                    'Middle Name': middlenamey.get().upper(),
                    'Last Name': lastnamey.get().upper(),
                    'Gender': Var.get().upper(),
                    'Email': emaily.get().upper(),
                    'Telephone No': tpy.get().upper(),
                    'Address': add1.get(),
                    'Course': course1.get(),
                    'Department': dept1.get(),
                    'Day of Registration':dd,
                    'Date of Registration':main_day,
                    'Image Path': f'{path1}'
                    }
                df = df.append(data, ignore_index=True)
                df.to_csv(file_name, index = False)
            else:
                data = {
                    'First Name': firstnamey.get(),
                    'Middle Name': middlenamey.get(),
                    'Last Name': lastnamey.get(),
                    'Gender': Var.get(),
                    'Email': emaily.get(),
                    'Telephone No': tpy.get(),
                    'Address': add1.get(),
                    'Course': course1.get(),
                    'Department': dept1.get(),
                    'Day of Registration':dd,
                    'Date of Registration':main_day,
                    'Image Path': f'{path1}'
                    }
                df = pd.DataFrame([data]).to_csv(file_name, index=False)
            # clear the interface once done
            firstnamey.delete(0, END)
            middlenamey.delete(0, END)
            lastnamey.delete(0,END)
            emaily.delete(0, END)
            tpy.delete(0,END)
            add1.delete(0,END)
            course1.delete(0,END)
            dept1.delete(0,END)
            firstnamey.focus()         
            messagebox.showinfo(profile, 'Your record has been saved', icon='info')
        else:
            messagebox.showerror(profile, 'Take picture before Sumbit',icon='warning')

    submit1 = Button(profile, text = 'Submit', command = submit,font=('Times', 12), bg= 'black', fg= 'blue')
    submit1.grid(row = 9, column=3)
    # the exit button on the menubar
    def exit():
        exitl = messagebox.askquestion(profile, 'Are you sure you want to exist?',icon='warning')  
        if exitl == 'yes':
            profile.destroy()
    exit1=Button(profile,text='Exit', font =('Times',10),command=exit,bg='black',fg='red')
    exit1.grid(row=10,column=3)

nu= Label (ws, text= ' ', font = ('Times', 15), fg = 'black',width =30)
nu.grid(row=0,column =3)
profile = Button(ws, text ='Create a Profile', width = 15,command = register, font = ('Times', 15), bg='black', fg='orange')    
profile.grid(row=1, column = 3)

def mark_att():
    from datetime import datetime,date
    mark = Toplevel(ws)
    mark.title('Mark Student Attendance')
    mark.geometry('500x600')
            
    menubar = Menu(ws)
    mark.config(menu=menubar)
    homemenu= Menu(menubar,tearoff=0)
    homemenu.add_command(command=mark.destroy,label='Exit')
    menubar.add_cascade(label="File",menu=homemenu)

   
    filename = 'Student_Attendance.csv'
    day = pd.Timestamp(datetime.today())
    day= day.day_name()
    today = date.today()
    date = today.strftime("%m/%d/%y")
    date_string = date
    datetime = datetime.strptime(date_string, '%m/%d/%y')

    messagebox.showinfo(mark, f'Time to mark today attendance {day},{date} \n Press S to end marking of Attendance', icon='info')
    with open ('Student_encoder_new','rb') as f: # loading the pickle 
        saved_encodings = pickle.loads(f.read())
    stu_name = []
    video = cv2.VideoCapture(0)
    while 1:
        ret,frame = video.read()
        rgb_frame = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
        face = face_recognition.face_locations(rgb_frame,model='hog')
        video_encoding = face_recognition.face_encodings(rgb_frame,face)
        recognized_faces =[]
        cont = 1
        for encoding in video_encoding:
            distance = face_recognition.face_distance(encoding,saved_encodings['Encodings']) # bring the new face and the one in the pickle file together
            min_distance = np.min(distance) # geeting the minimum value from both 
            if min_distance > 0.5: # if the min value is greater than 0.5, then that means the face is not in our database 
                recognized_faces.append('Face not recognized')
                break
            else:
                index = np.argmin(distance) # get the highest number from both 
                recognized_faces.append(saved_encodings['Names'][index]) # get the name and the highest number
                for (top,right,bottom,left),reg_name in zip(face,recognized_faces): # zip the face in the database and the one that is highly correlated in the database together 
                    frame = cv2.rectangle(frame,(left,top),(right,bottom),(126,120,186),1) # form rectangler box 
                    frame = cv2.putText(frame,reg_name,(left,top -3),cv2.FONT_HERSHEY_COMPLEX,1,(255,10,125),1) # get the name
                    stu_name.append(reg_name) # apppend the name gotten into the empty list as they are marking the attendance 
                    number_of_student = set(stu_name)
                    total_num = len(number_of_student)
                    att = 'Present'

        cv2.imshow('face_recognition',frame)
        key = cv2.waitKey(10)
        if key == ord('s'):
            break
    video.release()
    cv2.destroyAllWindows()
    # create a database for the attandance and fill those that did not mark the attenance as absent
    if os.path.exists(filename):
        df = pd.read_csv(filename)
        df = df.fillna('Absent')
        data = {
            'Date': datetime,
            'Day': day,
            reg_name : att,
            }
        df = df.append(data, ignore_index=True)
        df.to_csv(filename, index = False)
    else:
        data = {
            'Date': datetime,
            'Day':day,
            reg_name : att,
             }
        df = pd.DataFrame([data]).to_csv(filename, index=False)
    messagebox.showinfo(mark, f'{total_num} student attendance been marked for today as at {day},{date}', icon='info')

label= Label(ws, text ='', font = ('Times', 15),  fg = 'black')    
label.grid(row=2, column = 3)    
mark_att_but =Button(ws, text ='Mark Attendance', width = 15,command = mark_att, font = ('Times', 15), bg='black', fg='orange')
mark_att_but.grid(row=3, column = 3)

def student_info():
    stu_info = Toplevel(ws)
    stu_info.title('Student Info')
    stu_info.geometry('500x600')
            
    menubar = Menu(ws)
    stu_info.config(menu=menubar)
    homemenu= Menu(menubar,tearoff=0)
    homemenu.add_command(command=stu_info.destroy,label='Exit')
    menubar.add_cascade(label="File",menu=homemenu)
    
    # building a tree to display the student details
    TableMargin = Frame(stu_info, width=500)
    TableMargin.pack(side=BOTTOM)
    scrollbarx = Scrollbar(TableMargin, orient=HORIZONTAL)
    scrollbary = Scrollbar(TableMargin, orient=VERTICAL)
    #column names needed for the table
    tree = ttk.Treeview(TableMargin, columns=('S/N','First Name', 'Middle Name', 'Last Name', 'Gender', 'Email','Telephone No', 'Address', 'Course', 'Department','Day of Registration', 'Date of Registration', 'Image Path'), height=400, selectmode="extended", yscrollcommand=scrollbary.set, xscrollcommand=scrollbarx.set,style='PViewStyle.Treeview')
    scrollbary.config(command=tree.yview)
    scrollbary.pack(side=RIGHT, fill=Y)
    scrollbarx.config(command=tree.xview)
    scrollbarx.pack(side=BOTTOM, fill=X)
    # column headings needed for the table
    tree.heading("S/N", text="S/N", anchor=W)
    tree.heading("First Name", text="First Name", anchor=W)
    tree.heading("Middle Name", text="Middle Name", anchor=W)
    tree.heading("Last Name", text="Last Name", anchor=W)
    tree.heading("Gender", text="Gender", anchor=W)
    tree.heading("Email", text="Email", anchor=W)
    tree.heading("Telephone No", text="Telephone No", anchor=W)
    tree.heading("Address", text="Address", anchor=W)
    tree.heading("Course", text="Course", anchor=W)
    tree.heading("Department", text="Department", anchor=W)
    tree.heading("Day of Registration", text="Day of Registration", anchor=W)
    tree.heading("Date of Registration", text="Date of Registration", anchor=W)
    tree.heading("Image Path", text="Image Path", anchor=W)
    # index of each column name 
    tree.column('#0', stretch=NO, minwidth=0, width=0) 
    tree.column('#1', stretch=NO, minwidth=0, width=0)
    tree.column('#2', stretch=NO, minwidth=0, width=200)
    tree.column('#3', stretch=NO, minwidth=0, width=200)
    tree.column('#4', stretch=NO, minwidth=0, width=200)
    tree.column('#5', stretch=NO, minwidth=0, width=200)
    tree.column('#6', stretch=NO, minwidth=0, width=200)
    tree.column('#7', stretch=NO, minwidth=0, width=200)
    tree.column('#8', stretch=NO, minwidth=0, width=200)
    tree.column('#9', stretch=NO, minwidth=0, width=200)
    tree.column('#10', stretch=NO, minwidth=0, width=200)
    tree.pack()
    
    
    sn = []
    counter = 1
    sn.append(counter)
    counter +=1
    with open("Student Profile.csv",encoding="utf8",errors="ignore",newline="") as f: #reading the database of the student profile
        lines = sum (1 for _ in csv.DictReader(f, delimiter=',')) 
        reader = csv.DictReader(f, delimiter=',')
        f.seek(0)
        # getting each column from the database
        for index, row in enumerate(reader, 1) :
            first = row["First Name"]
            middle = row['Middle Name']
            last = row['Last Name']
            gender = row['Gender']
            email = row['Email']
            tel = row['Telephone No']
            add = row['Address']
            cou = row['Course']
            dept = row['Department']
            dr = row['Day of Registration']
            date = row['Date of Registration']
            im = row['Image Path']
            tree.insert("", 0, values=(lines-index+1,first,middle,last,gender,email,tel,add,cou,dept,dr,date,im))

label= Label(ws, text ='', font = ('Times', 15),  fg = 'black')    
label.grid(row=6, column = 3)    
mark_att_but =Button(ws, text ='Student Details', width = 15,command =student_info, font = ('Times', 15), bg='black', fg='orange')
mark_att_but.grid(row=7, column = 3)


ws.mainloop()
```
